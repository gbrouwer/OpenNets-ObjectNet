### Divide Training / Validation Ratio

```python
divideRatio = 0.8
```

### Find the available files

```python
#Find Available Files for Empty Patches
objects = []
for root, dirs, files in os.walk("../../data/objectNet/Uganda/raw/structure/", topdown=False):
    for name in files:
        objects.append(os.path.join(root, name))
        
controls = []
for root, dirs, files in os.walk("../../data/objectNet/Uganda/raw/control/", topdown=False):
    for name in files:
        controls.append(os.path.join(root, name))        
```

### Divide into train and validation

```python
#What is the smallest category?
nRaw = np.min([len(objects),len(controls)])
nRaw = 8000
```

```python
#Create a random permuration of the images
randindices = np.random.permutation(nRaw)
divider = int(np.round(nRaw*divideRatio))
trainIndices = randindices[:divider]
valIndices = randindices[divider:]
```

```python
print trainIndices
```

```python
#Delete Previous Data and create directories
os.system('rm -rf ../../data/objectNet/Uganda/train/structure');
os.system('rm -rf ../../data/objectNet/Uganda/val/structure');
os.system('rm -rf ../../data/objectNet/Uganda/train/control');
os.system('rm -rf ../../data/objectNet/Uganda/val/control');
os.system('mkdir -p ../../data/objectNet/Uganda/val/structure');
os.system('mkdir -p ../../data/objectNet/Uganda/train/structure');
os.system('mkdir -p ../../data/objectNet/Uganda/val/control')
os.system('mkdir -p ../../data/objectNet/Uganda/train/control');
```

```python
#Copy train indices images over
for i in tqdm(range(0,len(trainIndices),1)):
    index = trainIndices[i]
    orig = controls[index]
    dest = orig.replace('raw','train')
    os.system('cp ' + orig + ' ' + dest)
    orig = objects[index]
    dest = orig.replace('raw','train')
    os.system('cp ' + orig + ' ' + dest)    
```

```python
#Copy val indices images over
for i in tqdm(range(0,len(valIndices),1)):
    index = valIndices[i]
    orig = controls[index]
    dest = orig.replace('raw','val')
    os.system('cp ' + orig + ' ' + dest)
    orig = objects[index]
    dest = orig.replace('raw','val')
    os.system('cp ' + orig + ' ' + dest)    
```
